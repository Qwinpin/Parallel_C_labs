# Quick

```python
axis = [i for i in range(1000, 10000000, 100000)]
```

```python
res = []
with open('quick_process.txt') as f:
    for line in f:
        res.append(line)
f.closed
```

```python
quick_sort = [[float(j) for j in res[i][:-3].split(', ')] for i in range(1, len(res)) if i % 2 == 0]
```

```python
data2 = [pyplt.Scatter(
    x=axis,
    y=val,
    mode='lines',
    name='Num of threads {}'.format(i + 1)) for i, val in enumerate(quick_sort) if i % 2 == 0]

layout2 = dict(title = 'Compare sort algorithm',
    xaxis = dict(title = 'Elements number'),
    yaxis = dict(title = 'Time (ms)'))
```

```python
plotly.offline.iplot({
    'data': data2,
    'layout': layout2})
```
